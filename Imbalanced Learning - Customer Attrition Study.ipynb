```python
def plot_woe(res, x, iv_dict, pv_dict, cv_dict):
    '''
    Positive WOE implies higher probability of event 1 vs. 0
    Negative WOE  the opposite.
    '''
    res.columns = ['count', 'ones']
    total_count, total_ones = res.sum()
    total_zeros = total_count - total_ones
    res['zeros'] = res['count'] - res['ones']
    res['zeros_pct'] = res['zeros']/total_zeros
    res['ones_pct']  = res['ones']/total_ones
    
    # Calculating WOE & Information Value
    res['woe'] = np.log(res['ones_pct']/res['zeros_pct'])
    res['iv']  = res['woe']*(res['ones_pct'] - res['zeros_pct'])
    
    # Calculating other statistics
    chi = chi2_contingency(np.array(res[['zeros', 'ones']]))
    n   = res['count'].sum()
    cv  = np.sqrt(chi[0]/n)
    
    # Updating dicts
    iv_dict[x] = res['iv'].sum()
    pv_dict[x] = chi[1]
    cv_dict[x] = cv
    
    sns.barplot(data=res, x=list(res.index), y='woe', color='grey').set_xlabel(x)
    plt.xticks(rotation=45)
    return iv_dict, pv_dict, cv_dict


def calculate_woe(dataframe, feats, label, ncol=3):
    '''
    df: dataframe
    feats: variable of interest
    label: output variable
    '''
    n, rows, cols = len(feats), int(np.ceil(len(feats)/ncol)), ncol
    plt.figure(figsize=(15, 3*rows))
    df = dataframe[feats+[label]].copy()
    iv_dict  = dict()      #information value
    pv_dict = dict()       #p-value
    cv_dict  = dict()      #cramers_v value
    
    for i, x in enumerate(feats):
        # Categorical feature
        if df[x].dtype == object:
            res = df[[x,label]].groupby(by=x).agg(['count', 'sum'])
            plt.subplot(rows, cols, i+1)
            iv_dict, pv_dict, cv_dict = plot_woe(res, x, iv_dict, pv_dict, cv_dict)

        # Numeric feature - try binning
        else:
            for bins_num in range(20, 1, -1):
                df['bin'] = pd.qcut(df[x], bins_num, duplicates='drop').apply(lambda x: x.right).astype(float)
                df_grouped = pd.DataFrame(df.groupby('bin').agg({x:'count', label: 'sum'})).reset_index()
                r = abs(spearmanr(df_grouped['bin'], df_grouped[label]/df_grouped[x])[0]) #/df[x] because of uneven size
                
                if (
                    r==1
                    and df_grouped[x].min() > .05*df_grouped[x].sum() # minimum 5% in each bucket
                    and not (df_grouped[x] == df_grouped[label]).any() # contains both events/ non-events
                    ):
                    bin_feasible = True
                    break

            # df['bin'] = df['bin'].fillna(999) # to include NaN in output
            res = df[['bin',label]].groupby(by='bin').agg(['count', 'sum'])
            plt.subplot(rows, cols, i+1)
            iv_dict, pv_dict, cv_dict = plot_woe(res, x, iv_dict, pv_dict, cv_dict)
    plt.tight_layout()
    
    return iv_dict, pv_dict, cv_dict
```

```python
def interpret_iv(x):
    if x<0.02:  return 'useless'
    elif x<0.1: return 'weak'
    elif x<0.3: return 'medium'
    elif x<0.5: return 'strong'
    else:       return 'very strong/suspicious'
    
def interpret_cv(x):
    if   x<0.1: return 'useless'
    elif x<0.2: return 'weak'
    elif x<0.4: return 'medium'
    elif x<0.6: return 'strong'
    else:       return 'very strong/suspicious'
```

```python
IV = pd.DataFrame(index=iv.keys(), data=iv.values())
IV.columns = ['Information_Value']
IV.sort_values(by='Information_Value', inplace=True, ascending=False)
IV['IV_Interpret'] = IV['Information_Value'].apply(interpret_iv)
# IV

PV = pd.DataFrame(index=pv.keys(), data=pv.values())
PV.columns = ['P_value']
PV['P_value'] = PV['P_value'].round(2)
# PV

CV = pd.DataFrame(index=cramers_v.keys(), data=cramers_v.values())
CV.columns = ['CramersV']
CV['CramersV'] = CV['CramersV'].round(2)
# CV
```

```python
IV_plot = IV[IV['Information_Value']>0.1]
plt.figure(figsize=(8,5))
sns.barplot(x=IV_plot.index, y=IV_plot['Information_Value'], palette='viridis')
plt.xticks(rotation=90);
```

```python
res = pd.concat([IV, PV, CV], axis=1, sort='False')
res['CV_Interpret'] = res['CramersV'].apply(interpret_cv)
res.sort_values(by='Information_Value', ascending=False)
```

```python
def get_data(df, att_group=[1,2], drop_feats=[], test_size=0.3, random_state=1, 
             smote=False, oversample=False, undersample=False, adasyn=False, mixed=False):
    '''
    att_group: attrited groups to include in analysis
    drop_feats: list of feature names to be dropped from X
    test_size, random_state: parameters for test_train_split
    '''
    
    # Dummy columns for categorical feature (known) 
    dummy_col = pd.get_dummies(df['hoh_segment_code'], drop_first=True)
    df_model = pd.concat([df.drop(columns=['hoh_segment_code']), dummy_col], axis=1)
    print('Shape of raw dataset: ', df_model.shape, '\n', df_model['attrition_flag'].value_counts())
    
    # Include selected Attrition Groups
    df_att_0 = df_model[df_model['att_group'].isin([0])]
    df_att_1 = df_model[df_model['att_group'].isin(att_group)]
    print('Shape of non-attrited & attrited dfs :', df_att_0.shape, df_att_1.shape)
    
    df_model = df_att_1.append(df_att_0)
    
    # Check for any missing values - should be none, because of prior cleanup
    print('Dropping missing rows if any: '), missing(df_model)       
    
    # Create X, y - drop unwanted features
    feats_to_drop = ['att_group', 'attrition_flag'] + drop_feats
    y = df_model['attrition_flag']
    X = df_model.drop(columns=feats_to_drop)
    print('Shape of dataset before under/over sampling: ', X.shape, '\n', y.value_counts())
    
    # Train test split
    X_tr, X_ts, y_tr, y_ts = train_test_split(X, y, test_size=test_size, random_state=random_state)
    
    # UNDER-SAMPLER
    if undersample:
        print('...running UNDER SAMPLER...')
        rus = RandomUnderSampler(random_state=0)
        X_tr, y_tr = rus.fit_resample(X_tr, y_tr)
    
    # SMOTE
    if smote:
        print('...running SMOTE...')
        sm = SMOTE(random_state=0)
        X_tr, y_tr = sm.fit_resample(X_tr, y_tr) 
    
    # ADASYN
    if adasyn:
        print('...running ADASYN...')
        ad = ADASYN(random_state=0)
        X_tr, y_tr = ad.fit_resample(X_tr, y_tr) 
    
    # OVER-SAMPLER
    if oversample:
        print('...running OVER SAMPLER...')
        ros = RandomOverSampler(random_state=0)
        X_tr, y_tr = ros.fit_resample(X_tr, y_tr)
    
    # MIXED SAMPLER = OVER + UNDER
    if mixed:
        over = SMOTE(random_state=0, sampling_strategy=0.5)
        under = RandomUnderSampler(random_state=0,sampling_strategy=1.0)
        pipeline = Pipeline([('over', over),('under', under)])
        X_tr, y_tr = pipeline.fit_resample(X_tr, y_tr)
    
    print('X_train: ', X_tr.shape, '\n', np.bincount(y_tr))
    print('X_test: ', X_ts.shape, '\n', y_ts.value_counts())
    

    # Return train test split datasets
    return X_tr, X_ts, y_tr, y_ts
```

```python
def run_model_summary(estimator, X_train, X_test, y_train, y_test, fit=True, lift=True):
    
    # Fit & Predict
    model = estimator
    fi = []
    
    if fit:
        print('...fitting model...')
        model.fit(X_train, y_train)
    
    print('...using model...')
    y_pred = model.predict(X_test)
    print('Original performance: \n', confusion_matrix(y_test, y_pred), '\n', classification_report(y_test, y_pred))
    print('Original F-score: {:.4f}'.format(f1_score(y_test, y_pred)))
    
    #print('Geometric mean: ', geometric_mean_score(y_test, y_pred))
    
    # Fscore to change threshold
    try:
        y_prob = model.predict_proba(X_test)
        y_prob = y_prob[:,1]
        precision, recall, thresholds = precision_recall_curve(y_test, y_prob)
        try:
            fscore = 2*precision*recall/(precision+recall)
            ix = fscore.argmax()
            cutoff = thresholds[ix]
        except:
            print('...failed to maximize F-score, changing default threshold to 0.7...')
            cutoff = 0.7
        y_pred_fs = [1 if i >= cutoff else 0 for i in y_prob]
        print('Revised threshold : {:.2f}'.format(cutoff))
        print('Revised F-score: {:.4f}'.format(f1_score(y_test, y_pred_fs)))
        print('Confusion matrix: \n', confusion_matrix(y_test, y_pred_fs))    
    except:
        pass
    
    # Feature Importance
    try:
        fi = pd.Series(model.feature_importances_, index = X_test.columns)
        fi = fi.sort_values(ascending = False)
        fi_gt01 = fi[fi>0.01]
    except:
        pass
    
    plt.figure(figsize = (20,8))
    
    # Plot Feature Importance
    try:
        plt.subplot(1,2,1)
        sns.barplot(x = fi_gt01.index, y = fi_gt01.values, color='blue');
        plt.xticks(rotation=90);
    except:
        pass
    
    # Plot Precision-Recall curve
    try:
        plt.subplot(1,2,2)
        sns.lineplot(x=recall, y=precision, markers='#').set(xlabel='Recall', ylabel='Precision')
        sns.scatterplot(x=[recall[ix]], y=[precision[ix]], color='black', marker='o')
    except:
        pass
    
    # Print Lift Table:
    
    if lift:
        try:
            print('Lift table: \n')
            y_prob = pd.Series(y_prob)
            lift = pd.DataFrame(pd.concat([y_prob.reset_index(drop=True), y_test.reset_index(drop=True)], axis=1, ignore_index=False))
            lift.columns = ['prob', 'label']
            mean = lift['label'].mean()
            lift['bin'] = lift['prob'].apply(lambda x: np.ceil(x*10)).astype(int)
            lift = lift.groupby(by='bin').agg({'prob':'mean', 'label':['count', 'mean']})
            lift[('label', 'lift')] = round(lift[('label', 'mean')]/mean,2)
            print(lift)
        except:
            pass
        
    return model, fi
```

```python
# KNN for anomaly detection

dummy_col = pd.get_dummies(df['hoh_segment_code'], drop_first=True)
df_model = pd.concat([df.drop(columns=['hoh_segment_code']), dummy_col], axis=1)
print('Shape of raw dataset: ', df_model.shape, '\n', df_model['attrition_flag'].value_counts())
    
# Include selected Attrition Groups
feats_to_drop = ['att_group'] + ignore_feats2

df_model  = df_model[df_model['att_group'].isin([0,1,2])].drop(columns=feats_to_drop)
X = df_model.drop(columns='attrition_flag')
y = df_model['attrition_flag']
print(X.shape, y.shape)

nbrs = NearestNeighbors(n_neighbors=3)
nbrs.fit(X)

dist, indexes = nbrs.kneighbors(X)

dist_avg = dist.mean(axis=1)
print(dist.shape, len(dist_avg))

pt = [50, 60, 70, 80, 90, 95, 99]
thresholds = np.percentile(dist_avg, pt)
thresholds

for i in range(len(pt)):
    y_pred = dist_avg>thresholds[i]
    print(pt[i],' percentile distance threshold = {:.2f}'.format(thresholds[i]),'\n', confusion_matrix(y, y_pred))
```

### Neural Network

```python
def ANN(X_train, y_train, X_test, y_test, loss, weights, ep=50):
    model = keras.Sequential([
        keras.layers.Dense(42, input_dim=42, activation='relu'),
        keras.layers.Dense(15, activation='relu'),
        keras.layers.Dense(1, activation='sigmoid')
    ])
    
    model.compile(optimizer='adam', loss=loss, metrics=['accuracy'])
    
    if weights == -1:
        model.fit(X_train, y_train, epochs=ep)
    else:
        model.fit(X_train, y_train, epochs=ep, class_weight = weights)
        
    print(model.evaluate(X_test, y_test))
    
    y_pred = model.predict(X_test)
    y_pred = np.round(y_pred)
    
    print(classification_report(y_test, y_pred))
    print(confusion_matrix(y_test, y_pred))
    return model
```
